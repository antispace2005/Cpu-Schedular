```python

Processes = [
    # P1 arrives at the very beginning. With a long burst, it will definitely be preempted.
    Process(pid=1, arrival_time=0, burst_time=10, priority=2),
    
    # P2 and P3 arrive quickly after P1, stacking up the ready queue.
    Process(pid=2, arrival_time=1, burst_time=4, priority=1),
    Process(pid=3, arrival_time=2, burst_time=5, priority=3),
    
    # P4 has a very short burst. It's a good test to ensure your scheduler 
    # properly handles a process finishing before the time quantum expires.
    Process(pid=4, arrival_time=4, burst_time=2, priority=0),
    
    # P5 arrives exactly when a standard time quantum (like Q=4 or Q=5) might be triggering 
    # a context switch. This tests if your queue appends new arrivals before or after preempted processes.
    Process(pid=5, arrival_time=5, burst_time=8, priority=2),
    
    # P6 tests simultaneous arrival. If P5 gets preempted at t=10, and P6 arrives at t=10, 
    # your queue logic needs to decide who goes in first.
    Process(pid=6, arrival_time=10, burst_time=3, priority=1),
    
    # P7 tests CPU idle time. The CPU should completely drain the ready queue and sit idle 
    # for a while before this process arrives.
    Process(pid=7, arrival_time=50, burst_time=7, priority=3)
]
scheduler = RRscheduler(Processes,3)
scheduler.schedule()
scheduler.gantt_chart_array
```
